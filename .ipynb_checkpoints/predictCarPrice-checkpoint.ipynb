```python
import pandas as pd
data = pd.read_csv("./data/autos/largeDF.csv.gz",compression="gzip")
data = data[["Modell","Marke","Garantie","country","Zustand", "Zylinder","Kraftstoff","Erstzulassung", "Außenfarbe","Innenausstattung","Karosserieform","Anzahl Türen","Sitzplätze","Getriebeart","Gänge",'Hubraum', 'Kraftstoff', 'Schadstoffklasse', 'haendler', 'privat',"price"]]
data = data.fillna(0)
rawdata = data.copy()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
def preprocess(data):
    obj_df = data.select_dtypes(include=['object']).copy()
    num_df = data.select_dtypes(include=['float64']).copy()
    print(obj_df.columns)
    obj_df = pd.get_dummies(obj_df)
    scaler = MinMaxScaler()
    cols = num_df.columns
    
    num_df = scaler.fit_transform(num_df)
    num_df = pd.DataFrame(num_df,columns=cols)
    data = pd.concat([obj_df,num_df],axis=1)
    return data, scaler

data,scaler = preprocess(data)
```

```python
y = data["price"]
data = data.drop(["price"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
#x_train, x_test, y_train, y_test = train_test_split( data, y, test_size=0.33, random_state=42,shuffle=True)
# have to do other split for later

splitAt = int(len(data)*0.9)
x_train,x_test = data[:splitAt],data[splitAt:]
y_train,y_test = y[:splitAt],y[splitAt:]
```

```python
# best mdoel
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(criterion= "mse", max_depth= None, max_features= "sqrt", min_samples_split= 2, n_estimators= 500
)
```

```python
model.fit(x_train, y_train)
scr = model.score(x_test,y_test)
pred = model.predict(x_test)

# whole model
model.fit(data,y)
pred_whole = model.predict(data)
```

```python
import joblib
# save model
joblib.dump(model, "randomforest_fullytrained.sav", compress=1)
joblib.dump(scaler, "scaler_fullytrained.sav", compress=1)
```

```python
diffs = []

for pds,real in zip(pred,y_test):
    diff = real - pds
    diffs.append(diff)
    
print(diff.mean())
```

```python
import matplotlib.pyplot as plt

plt.boxplot(diffs)
```

```python
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(diffs))
threshold = 6
goodguesses = np.where(z < 0.1)
outliers = np.where(z > threshold)
print(outliers)

# prepare rawdata
rawdata_whole = rawdata.copy()
rawdata = rawdata[splitAt:]
```
